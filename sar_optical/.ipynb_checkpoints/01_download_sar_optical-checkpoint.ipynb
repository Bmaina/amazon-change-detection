# 🛰️ 01 — Download SAR + Optical Data
### Amazon Deforestation — Multi-Sensor Change Detection

Downloads both **Sentinel-1 SAR** (C-band, VV+VH polarisation) and
**Sentinel-2 optical** (RGB+NIR) for Rondônia, Brazil — 2019 and 2024.

**Why SAR + Optical fusion?**
- Optical imagery is blocked by clouds — Amazon has 70-80% cloud cover year-round
- SAR penetrates clouds and works day/night
- Fusing both gives cloud-robust deforestation detection
- This is the same methodology used by INPE (Brazil's deforestation agency) and ESA

**Pipeline:**
1. 🛰️ `01_download_sar_optical.ipynb` ← *you are here*
2. 🔧 `02_prepare_sar_optical.ipynb`
3. 🌿 `03_fusion_change_detection.ipynb`
4. 🎨 `04_visualise_fusion.ipynb`


## ⚙️ Configuration

```python
from pathlib import Path
import numpy as np

# ── Directories ───────────────────────────────────────────────────────────────
RAW_DIR = Path("data/raw_sar_optical")
RAW_DIR.mkdir(parents=True, exist_ok=True)

# ── Area of Interest — Rondônia fishbone deforestation ────────────────────────
BBOX = (-63.5, -11.5, -62.5, -10.5)   # lon_min, lat_min, lon_max, lat_max

# ── Date ranges ───────────────────────────────────────────────────────────────
DATE_RANGES = {
    "2019": ("2019-07-01", "2019-09-30"),
    "2024": ("2024-07-01", "2024-09-30"),
}

print("Configuration set")
print(f"  AOI : {BBOX}")
print(f"  2019: {DATE_RANGES['2019']}")
print(f"  2024: {DATE_RANGES['2024']}")
print(f"  Raw dir: {RAW_DIR.resolve()}")
```

## 📥 Step 1 — Download Sentinel-1 SAR

Sentinel-1 provides C-band SAR imagery (5.6cm wavelength) in VV and VH polarisations.
- **VV** (vertical-vertical): sensitive to surface roughness — good for bare soil vs forest
- **VH** (vertical-horizontal): sensitive to volume scattering — good for forest structure

**Free access** via Copernicus Data Space: [dataspace.copernicus.eu](https://dataspace.copernicus.eu)

**Manual download instructions:**
1. Go to [browser.dataspace.copernicus.eu](https://browser.dataspace.copernicus.eu)
2. Search: Sentinel-1, GRD, IW mode, VV+VH
3. Draw AOI over Rondônia: lon -63.5 to -62.5, lat -11.5 to -10.5
4. Date: 2019-07-01 to 2019-09-30, then 2024-07-01 to 2024-09-30
5. Download and extract to `data/raw_sar_optical/sar/2019/` and `data/raw_sar_optical/sar/2024/`


```python
import requests

STAC_URL = "https://catalogue.dataspace.copernicus.eu/resto/api/collections/Sentinel1/search.json"

def search_sentinel1(bbox, start, end):
    lon_min, lat_min, lon_max, lat_max = bbox
    params = {
        "startDate":       start,
        "completionDate":  end,
        "productType":     "GRD",
        "sensorMode":      "IW",
        "polarisation":    "VV%26VH",
        "box":             f"{lon_min},{lat_min},{lon_max},{lat_max}",
        "maxRecords":      3,
        "sortParam":       "startDate",
        "sortOrder":       "descending",
    }
    try:
        resp = requests.get(STAC_URL, params=params, timeout=30)
        resp.raise_for_status()
        features = resp.json().get("features", [])
        print(f"   Found {len(features)} Sentinel-1 scenes")
        for f in features:
            print(f"   - {f['properties'].get('title', 'unknown')}")
        return features
    except Exception as e:
        print(f"   Search failed: {e}")
        return []

print("Searching for Sentinel-1 SAR scenes...")
for year in ["2019", "2024"]:
    print(f"  {year}:")
    search_sentinel1(BBOX, *DATE_RANGES[year])
```

## 📥 Step 2 — Download Sentinel-2 Optical

Same as the main pipeline — Sentinel-2 L2A RGB+NIR bands.


```python
STAC_S2 = "https://catalogue.dataspace.copernicus.eu/resto/api/collections/Sentinel2/search.json"

def search_sentinel2(bbox, start, end):
    lon_min, lat_min, lon_max, lat_max = bbox
    params = {
        "startDate": start, "completionDate": end,
        "processingLevel": "S2MSI2A", "cloudCover": "[0,20]",
        "box": f"{lon_min},{lat_min},{lon_max},{lat_max}",
        "maxRecords": 3, "sortParam": "cloudCover", "sortOrder": "ascending",
    }
    try:
        resp = requests.get(STAC_S2, params=params, timeout=30)
        resp.raise_for_status()
        features = resp.json().get("features", [])
        print(f"   Found {len(features)} Sentinel-2 scenes")
        for f in features:
            cc = f['properties'].get('cloudCover', '?')
            print(f"   - {f['properties'].get('title','unknown')}  cloud={cc}%")
        return features
    except Exception as e:
        print(f"   Search failed: {e}")
        return []

print("Searching for Sentinel-2 optical scenes...")
for year in ["2019", "2024"]:
    print(f"  {year}:")
    search_sentinel2(BBOX, *DATE_RANGES[year])
```

## 🏗️ Step 3 — Generate Synthetic SAR + Optical Data

If real data is unavailable, generate physically-motivated synthetic data
that simulates the SAR backscatter signatures of Amazon land cover types.

**SAR backscatter physics:**
- Forest: high VH (volume scattering from canopy), moderate VV → VH/VV ratio ~0.3-0.4
- Bare soil/deforested: low VH, high VV → VH/VV ratio ~0.1-0.15
- Water: very low both → specular reflection away from sensor
- Pasture: moderate VV, low VH → ratio ~0.2

This is the same physical basis used by INPE's PRODES deforestation detection system.


```python
import rasterio
from rasterio.transform import from_bounds

def create_synthetic_sar(year, bbox, size=512):
    np.random.seed(42 if year == "2019" else 99)
    H, W = size, size
    lon_min, lat_min, lon_max, lat_max = bbox

    # Base backscatter — Amazon forest = high VH, moderate VV
    # Units: dB, typical values:
    #   Forest   : VV ~ -8 dB,  VH ~ -14 dB
    #   Bare soil: VV ~ -6 dB,  VH ~ -18 dB
    #   Water    : VV ~ -20 dB, VH ~ -25 dB

    vv = np.random.normal(-8.0, 1.5, (H, W)).astype(np.float32)
    vh = np.random.normal(-14.0, 2.0, (H, W)).astype(np.float32)

    if year == "2024":
        # Add fishbone deforestation pattern — bare soil signature
        for i in range(0, W, 40):
            vv[:, i:i+8] = np.random.normal(-6.0, 1.0, (H, 8))   # bare soil VV
            vh[:, i:i+8] = np.random.normal(-18.0, 1.5, (H, 8))  # bare soil VH
        # Random clearings
        np.random.seed(99)
        for _ in range(20):
            cy, cx = np.random.randint(50, H-50), np.random.randint(50, W-50)
            r = np.random.randint(15, 35)
            y, x = np.ogrid[:H, :W]
            mask = (y-cy)**2 + (x-cx)**2 <= r**2
            vv[mask] = np.random.normal(-6.5, 1.0)
            vh[mask] = np.random.normal(-18.5, 1.5)

    out_dir = RAW_DIR / "sar" / year
    out_dir.mkdir(parents=True, exist_ok=True)
    transform = from_bounds(lon_min, lat_min, lon_max, lat_max, W, H)
    profile = {
        "driver": "GTiff", "dtype": "float32", "width": W, "height": H,
        "count": 1, "crs": "EPSG:4326", "transform": transform, "compress": "lzw",
    }
    for band, name in [(vv, "VV"), (vh, "VH")]:
        with rasterio.open(out_dir / f"S1_{name}_{year}.tif", "w", **profile) as dst:
            dst.write(band[np.newaxis, :])
    print(f"   SAR {year}: VV mean={vv.mean():.2f} dB  VH mean={vh.mean():.2f} dB")
    return out_dir

def create_synthetic_optical(year, bbox, size=512):
    np.random.seed(42 if year == "2019" else 99)
    H, W = size, size
    lon_min, lat_min, lon_max, lat_max = bbox

    # Sentinel-2 reflectance (0-10000 scale)
    # Forest: high NIR, low Red → high NDVI
    if year == "2019":
        b04 = np.random.normal(600,  50,  (H, W)).astype(np.float32)  # Red
        b08 = np.random.normal(3500, 200, (H, W)).astype(np.float32)  # NIR
    else:
        b04 = np.random.normal(1200, 150, (H, W)).astype(np.float32)
        b08 = np.random.normal(2200, 300, (H, W)).astype(np.float32)
        # Fishbone deforestation
        for i in range(0, W, 40):
            b04[:, i:i+8] = np.random.normal(2500, 200, (H, 8))
            b08[:, i:i+8] = np.random.normal(1500, 200, (H, 8))

    b03 = (b04 * 1.1 + np.random.normal(0, 30, (H,W))).astype(np.float32)  # Green
    b02 = (b04 * 0.8 + np.random.normal(0, 20, (H,W))).astype(np.float32)  # Blue

    out_dir = RAW_DIR / "optical" / year
    out_dir.mkdir(parents=True, exist_ok=True)
    transform = from_bounds(lon_min, lat_min, lon_max, lat_max, W, H)
    profile = {
        "driver": "GTiff", "dtype": "float32", "width": W, "height": H,
        "count": 1, "crs": "EPSG:4326", "transform": transform, "compress": "lzw",
    }
    for band, name in [(b04,"B04"), (b03,"B03"), (b02,"B02"), (b08,"B08")]:
        with rasterio.open(out_dir / f"S2_{name}_{year}.tif", "w", **profile) as dst:
            dst.write(band[np.newaxis, :])

    ndvi = (b08 - b04) / (b08 + b04 + 1e-6)
    print(f"   Optical {year}: NDVI mean={ndvi.mean():.3f}")
    return out_dir

print("Generating synthetic SAR + Optical data...")
print()
for year in ["2019", "2024"]:
    print(f"  {year}:")
    create_synthetic_sar(year, BBOX)
    create_synthetic_optical(year, BBOX)
    print()

print("Done. Proceed to: 02_prepare_sar_optical.ipynb")
```
